```python
This common "error" happens because your local Git project and the GitHub repository have different histories. When you created the README file on GitHub.com, GitHub created a commit that your local computer doesn't have yet.

Git refuses to push because it doesn't want you to accidentally overwrite that README file. You must first "pull" that commit down to your computer, combine it with your files, and then push everything back.

The Solution: Pull, then Push
Follow these steps in your Ubuntu terminal:

1. Commit your local changes first
Git won't let you pull if you have unsaved changes. Make sure your new files are committed locally:

Bash

#### git add .
#### git commit -m "Added my project files"
```

```python
2. Pull the changes from GitHub
Use the --rebase flag. This is the "cleanest" way for beginners because it takes your local work and neatly places it on top of the 
README file you created online.

Bash

#### git pull origin main --rebase
```

```python
3. Push back to GitHub
Now that your local history includes the README, you can push successfully:

Bash

#### git push origin main
Why did this happen?
```

Since there are now two different versions of "Commit #1," Git sees them as "diverged."

* git pull = Downloads the README and tries to merge it with your files.
* git push = Uploads the combined result back to the cloud.

```python
If the pull command gives you an error about "unrelated histories," use this specific command to force them to join:

Bash

#### git pull origin main --allow-unrelated-histories
```
