PySpark provides three modes for handling malformed records when reading data, particularly useful when the data does not perfectly conform to the specified schema:
* PERMISSIVE (Default Mode):
This mode inserts null values for fields that cannot be parsed correctly according to the defined schema.
If an entire record is malformed, it will be placed into a special column named _corrupt_record by default, 
while other fields in that row will be set to null.
This mode is suitable when you want to retain all records, including the malformed ones, and then analyze or clean them later.
* DROPMALFORMED:
This mode discards any records that contain fields that cannot be parsed correctly or do not conform to the schema.
Only records that are entirely valid and parseable according to the schema will be included in the resulting DataFrame.
This mode is appropriate when data integrity is paramount, and you prefer to work only with clean, valid data, even if it means losing some records.
* FAILFAST:
This mode immediately terminates the job and throws an exception as soon as it encounters a malformed record.
It is the strictest mode and is used when any data corruption is unacceptable, and the process should fail instantly upon detection.
This mode is valuable in scenarios where data quality is critical, and you need to ensure that no malformed data is processed, such as in financial applications or when loading data into sensitive systems.
