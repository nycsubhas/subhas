* To Create an udf, you must bring udf from pyspark.sql.functions import udf and then from pyspark.sql.types import StringType/IntegerType etc. 

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("udf").getOrCreate()
```

```python
# Create a DataFrame
columns = ["Seqno","Name"]
data = [(1, "john jones"),
    (2, "tracey smith"),
    (3, "amy sanders")]

df = spark.createDataFrame(data=data,schema=columns)

df.show(truncate=False)
```

```python
# Create a python Function 
def convertCase(str):
    resStr=""
    arr = str.split(" ")
    for x in arr:
       resStr= resStr + x[0:1].upper() + x[1:len(x)] + " "
    return resStr 
```

```python
from pyspark.sql.functions import col, udf
from pyspark.sql.types import *

# Converting function to UDF 
convertUDF = udf(lambda z: convertCase(z),StringType())  
```

```python
# Now apply convertUDF to the column
df2 = df.select("Seqno", convertUDF(col("Name")).alias("Name"))
df2.show()
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
# Create a DataFrame
columns = ["Seqno","Name"]
data = [(1, "john jones"),
    (2, "tracey smith"),
    (3, "amy sanders")]

df = spark.createDataFrame(data=data,schema=columns)

df.show(truncate=False)
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import IntegerType
@udf(IntegerType())
def double(x):
    return x*2
```

```python
df.select("Seqno", "name", double("Seqno").alias("double_seqno")).show()
```

```python
@udf(StringType())
def convertCase(str):
    resStr=""
    arr = str.split(" ")
    for x in arr:
       resStr= resStr + x[0:1].upper() + x[1:len(x)] + " "
    return resStr 
```

```python
df.show()
```

```python
df.select(convertCase("Name").alias("NAME").show()
```
