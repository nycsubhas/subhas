```python
from pyspark.sql import SparkSession 
spark = SparkSession.builder.appName("from_json").getOrCreate()
```

```python
path="/home/user/data/iot/iot_1.json"
df = (spark.read
      .format("text")
      .option("mode","permissive")
      .load(path) )
df.printSchema()
```

```python
from pyspark.sql.functions import from_json
from pyspark.sql.types import *
schema = StructType([
    StructField("id", IntegerType()),
    StructField("event_time", TimestampType()),
    StructField("value", IntegerType())
])
```

```python
parsed_df = df.select(from_json("value",schema).alias("value_struct"))
parsed_df.printSchema()
```

```python
# Now get back the actual columns which are inside the dataframe using "value_struct.*". 
final_df = parsed_df.select("value_struct.*")
final_df.printSchema()
```

```python
final_df.show()
```

* Identify the column which has to be parsed
* cast the column type into string (df.value.cast("string"))
* find the schema 
* Apply from_json: df.select(from_json(df.value, schema).alias("value_struct") --> parsed
* Now value_struct is of type "struct".
* Now do: parsed_df.select("value_struct.*") ---> final_df
* final_df.printSchema()/final_df.show()

```python
Thus, from_json parsed the column from string type into struct type. 
```
